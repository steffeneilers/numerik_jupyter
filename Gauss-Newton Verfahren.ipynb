```python
import sympy as sy
import numpy as np

P0 = [1, 0.5]
P1 = [2, -1.5]
P2 = [3, 3.5]
P3 = [4, -9]

lamda0 = [[2],[1]] #Startvektor
Pges = [P0, P1, P2, P3] #Alle Punkte eintragen
tolerance = 0.0000000000002
def f(x, l1, l2):
    return sy.sin(l1*x) * sy.exp(l2*x)



i = 0
lTuple = sy.symbols('l(0:len(Pges))')
x = sy.symbols('x')
while (True):
    fLambda = np.empty([len(Pges), 1])
    for j in range(0,len(Pges)):
        try: 
            fLambda[j][0]
            fLambda[j][0] = Pges[j][1] - f(Pges[j][0], lamda0[0][0], lamda0[1][0]).evalf()
        except AttributeError: 
            fLambda[j][0] = Pges[j][1] - f(Pges[j][0], lamda0[0][0], lamda0[1][0])

    jacobi = np.empty([len(Pges), 2])
    for j in range(0,len(Pges)):
        f_l1 = sy.diff(f(Pges[j][0], lTuple[0], lTuple[1]), lTuple[0])
        f_l2 = sy.diff(f(Pges[j][0], lTuple[0], lTuple[1]), lTuple[1])
        try: 
            jacobi[j][0] = f_l1.subs([(lTuple[0], lamda0[0][0]), (lTuple[1], lamda0[1][0])]).evalf()
            jacobi[j][1] = f_l2.subs([(lTuple[0], lamda0[0][0]), (lTuple[1], lamda0[1][0])]).evalf()
        except AttributeError: 
            jacobi[j][0] = f_l1.subs([(lTuple[0], lamda0[0][0]), (lTuple[1], lamda0[1][0])])
            jacobi[j][1] = f_l2.subs([(lTuple[0], lamda0[0][0]), (lTuple[1], lamda0[1][0])])

    if(np.linalg.norm(np.matmul(np.transpose(jacobi), fLambda)) < tolerance):
        raise ValueError("Genauigkeit erreicht")
    else:
        jacobiT = np.transpose(jacobi)
        d = np.matmul(np.matmul(np.linalg.inv(np.matmul(jacobiT, jacobi)), jacobiT), fLambda)
        lamda0 = lamda0 + d
        i+= 1
        print("\u03BB %d: \n" %i, lamda0)
```
