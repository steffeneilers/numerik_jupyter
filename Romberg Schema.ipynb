```python
import numpy as np
import matplotlib.pyplot as plt

#Intervallgrenzen [a,b]
a = 1
b = 3
n = 4 #Tiefe
def f(x):
    return 7*x*np.exp(np.sin(2*x))

t = np.zeros([b-a + abs(n-a), b-a + abs(n-a)]) #Matrix
t[1][1] = (b-a)/2*(f(a)+f(b))
for i in range (2,n+1):
    s = 0
    for k in range (1,2**(i-1),2):
        s += f(a + k * ((b-a)/(2**(i-1))))
    t[i][1] = 0.5 * (t[i-1][1]+((b-a)/(2**(i-2)))*s)
for i in range (2,n+1):
    for j in range (2,n+1):
        t[i][j] = (4**(j-1)*t[i][j-1]-t[i-1][j-1])/(4**(j-1)-1)
print(t[1:,1:], end="\n\n")

#Visual Output
print("Visual Output")
x = np.arange(a,b,(b-a)/50)
y = f(x)
plt.plot(x,y)
plt.show()
```
