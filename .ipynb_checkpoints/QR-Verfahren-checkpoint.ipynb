```python
### QR-Transformation ###

import numpy as np

maxiter=1
a=np.array([[4,-8,7,1],[-8,-2,-2,5],[7,-2,26,-1],[1,5,-1,4]])

def qrzerlegung(a):
    q=np.eye(len(a))
    for i in range (0,len(a)-1):
        for j in range (i+1,len(a)):
            if a[j][i]!=0:
                cos=a[i][i]/np.sqrt((a[j][i])**2+(a[i][i])**2)
                sin=a[j][i]/np.sqrt((a[j][i])**2+(a[i][i])**2)
                u=np.eye(len(a))  
                u[i][i]=cos
                u[i][j]=-sin
                u[j][i]=sin
                u[j][j]=cos
                ut=np.transpose(u)
                #print("uT",i+1,j+1,"=\n",ut)
                a=np.dot(ut,a)
                q=np.dot(q,u)
                #print("q=\n",q)
                #print("a=\n",a)
    return q , a

for i in range (0,maxiter):
    q,a= qrzerlegung(a)
    a=np.dot(a,q)
    print("Iterarion",i+1,"\n","a=\n",a,"\n","q=\n",q)
```
