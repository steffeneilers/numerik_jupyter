```python
import numpy as np

A = np.array([[9,0,-1],[0,2,1],[-1,1,5]])
b = np.array([[23],[0],[15]]) 
x0 = np.array([[0],[0],[0]]) #Startvektor
tolerance = 0.1 #Toleranz kleiner für höhere Genauigkeit

i = 0
while (True):
    d = b - np.matmul(A, x0)
    print("d%d: \n" % i, d)
    if(np.linalg.norm(d)< tolerance):
        raise ValueError("Genauigkeit erreicht")
    else:
        alpha = np.matmul(np.transpose(d), d) / np.matmul(np.matmul(np.transpose(d), A), d)
        print("\u03B1%d: \n" %i, alpha)
        x0 = x0 + np.multiply(alpha, d)
        print("x%d: \n" % (i+1), x0, end="\n----------------------\n")
        i+= 1
```

Konjugiertes-Gradienten-Verfahren

```python
import numpy as np

A = np.array([[3,-1],[2,-4]])
b = np.array([[1],[16]]) 
x0 = np.array([[0],[0]]) #Startvektor
tolerance = 0.01 #Toleranz kleiner für höhere Genauigkeit

r = b - np.matmul(A, x0)
d = r
i = 0
while (True):
    alpha = np.matmul(np.transpose(r), r) / np.matmul(np.transpose(d), np.matmul(A, d))
    xNext = x0 + np.multiply(alpha, d)
    rNext = r - np.multiply(alpha, np.matmul(A, d))
    beta = np.matmul(np.transpose(rNext), rNext) / np.matmul(np.transpose(r), r)
    d = rNext + np.multiply(beta, d)
    print("x%d: \n" % (i+1), xNext, end="\n----------------------\n")
    x0 = xNext
    r = rNext
    i+= 1
    if(np.linalg.norm(rNext) < tolerance):
        raise ValueError("Genauigkeit erreicht")
```
